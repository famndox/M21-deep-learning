## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(['EIN','NAME'], axis=1)
application_df.head()
```

```python
# Determine the number of unique values in each column.
application_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts to identify and replace with "Other"
app_type_summ = application_df['APPLICATION_TYPE'].value_counts()
app_type_summ
```

```python
# Attempting to emulate what I did in excel ...
df_app_type = pd.DataFrame(app_type_summ)
total_type = df_app_type['count'].sum()
df_app_type['cumulative_count'] = df_app_type['count'].cumsum()
df_app_type['cumulative_percentage'] = df_app_type['cumulative_count'] / total_type
df_app_type = df_app_type.drop(columns=['cumulative_count'])
df_app_type

# Based on the cumulative_percentage, I'm setting the cutoff at 500 to cover 99% of the logged application types
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
cutoff_app = 500
application_types_to_replace = list(app_type_summ[app_type_summ < cutoff_app].index)

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure replacement was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts to identify and replace with "Other"
class_summ = application_df['CLASSIFICATION'].value_counts()
class_summ
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
df_class_summ = pd.DataFrame(class_summ)
total_class = df_class_summ['count'].sum()
df_class_summ['cumulative_count'] = df_class_summ['count'].cumsum()
df_class_summ['cumulative_percentage'] = df_class_summ['cumulative_count'] / total_class
df_class_summ = df_class_summ.drop(columns=['cumulative_count'])
df_class_summ[:20]
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
cutoff_class = 33
classifications_to_replace = list(class_summ[class_summ < cutoff_class].index)

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"C-Other")

# Check to make sure replacement was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
df_got_dummies = pd.get_dummies(application_df)
df_got_dummies
```

```python
# Split our preprocessed data into our features and target arrays
X = df_got_dummies.drop(['IS_SUCCESSFUL'], axis=1).values
y = df_got_dummies['IS_SUCCESSFUL'].values
# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 1)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
inputs = len(X_train[0])
dlayer_1 = 8
dlayer_2 = 3

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=dlayer_1, activation="relu", input_dim=inputs)) 

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=dlayer_2, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_nn = nn.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
history_df = pd.DataFrame(fit_nn.history)

# Increase the index by 1 to match the number of epochs
history_df.index += 1

# Plot the loss
history_df.plot(y="loss")
```

```python
# Plot the accuracy
history_df.plot(y="accuracy")
```
